```python
from datetime import datetime, timezone
from facebook_scraper import get_posts, set_cookies, exceptions
import time
import pandas as pd
import json
```

```python
set_cookies("cookies.json")
```

```python
def handle_pagination_url(url):
    global start_url
    start_url = url
```

```python
start_url = None
```

```python
results = []
i = 0
start_time = datetime.now()
while True:
    try :
        for post in get_posts(group = "373920943948661", page_limit = 1,start_url = start_url, extra_info=False,
        request_url_callback = handle_pagination_url, options={"posts_per_page":1,"allow_extra_requests":True,"comments":True,"progress":True}):
            difftime = (datetime.now() - start_time).total_seconds()
            test_df = pd.DataFrame.from_dict(post, orient="index")
            # write every 10 minutes
            if (difftime%600)==0:
                now = datetime.now()
                nowtime = "{}-{}-{}_{}-{}-{}".format(now.year,now.month,now.day,now.hour,now.minute,now.second)
                print("## {} ## 10 minutes elapsed".format(nowtime))
            test_df = pd.DataFrame.from_dict(post, orient="index")
        print("Finished")
        break
    except exceptions.TemporarilyBanned:
        now = datetime.now()
        nowtime = "{}-{}-{}_{}-{}-{}".format(now.year,now.month,now.day,now.hour,now.minute,now.second)
        print("{} ## TEMPORARY BAN... sleeping for 20 min".format(nowtime))
        print(start_url)
        results = []
        time.sleep(1200)
```

```python
all_posts = pd.DataFrame()
for each in results:
    all_posts = all_posts.append(each, ignore_index = True)
if len(all_posts)>0:
    all_posts.to_csv("fb_scraper_{}.csv".format(nowtime))
    print("{} ## File written with timestamp".format(nowtime))
```
